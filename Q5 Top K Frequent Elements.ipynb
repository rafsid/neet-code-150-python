### Description

Top K Frequent Elements
Medium
Topics
Company Tags
Hints
Given an integer array nums and an integer k, return the k most frequent elements within the array.

The test cases are generated such that the answer is always unique.

You may return the output in any order.

Example 1:

Input: nums = [1,2,2,3,3,3], k = 2

Output: [2,3]
Example 2:

Input: nums = [7,7], k = 1

Output: [7]
Constraints:

1 <= nums.length <= 10^4.
-1000 <= nums[i] <= 1000
1 <= k <= number of distinct elements in nums.


Topics

       

Recommended Time & Space Complexity
You should aim for a solution with O(n) time and O(n) space, where n is the size of the input array.


Hint 1
A naive solution would be to count the frequency of each number and then sort the array based on each element’s frequency. After that, we would select the top k frequent elements. This would be an O(nlogn) solution. Though this solution is acceptable, can you think of a better way?


Hint 2
Can you think of an algorithm which involves grouping numbers based on their frequency?


Hint 3
Use the bucket sort algorithm to create n buckets, grouping numbers based on their frequencies from 1 to n. Then, pick the top k numbers from the buckets, starting from n down to 1.

### Code

#### O(n log n) Approach

```python
from typing import List

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        map = {}

        for j in nums:
            if j in map:
                map[j]+=1
            else:
                map[j]=1
        sorted_map = sorted(map.items(), key=lambda item: item[1], reverse=True)
        return [item[0] for item in sorted_map[:k]]
```

```python
nums = [1,2,2,3,3,3]
k =2 

solve = Solution()

solve.topKFrequent(nums,k)
```

```python
nums = [7,7]
k = 1

solve = Solution()

solve.topKFrequent(nums,k)
```

#### O(n) Approach

```python
from typing import List

class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        # Step 1: Count frequencies - O(n)
        freq_map = {}
        for num in nums:
            freq_map[num] = freq_map.get(num, 0) + 1
        
        # Step 2: Create buckets - O(n)
        # buckets[i] = list of numbers that appear i times
        buckets = [[] for _ in range(len(nums) + 1)]
        for num, freq in freq_map.items():
            buckets[freq].append(num)
        
        # Step 3: Collect top k from highest frequency - O(n)
        result = []
        for i in range(len(buckets) - 1, -1, -1):
            for num in buckets[i]:
                result.append(num)
                if len(result) == k:
                    return result
        
        return result
```

```python
nums = [1,2,2,3,3,3]
k =2 

solve = Solution()

solve.topKFrequent(nums,k)
```

```python
nums = [7,7]
k = 1

solve = Solution()

solve.topKFrequent(nums,k)
```
